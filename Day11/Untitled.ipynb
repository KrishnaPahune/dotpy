# matplotlib

```python
import matplotlib.pyplot as plt
x=[1,3,6,8]
y=[2,6,12,16]
plt.plot(x,y)
plt.show()
```

```python
x=[1,4,7]
y=[2,5,8]
x1=[1,2,3]
y1=[4,5,6]
plt.plot(x,y,label="Temperature")
plt.plot(x1,y1,label="humidity")
plt.xlabel("Trees")
plt.ylabel("Oxygen")
plt.legend()
plt.show()
```

```python
x=[2003,2004,2010]
y=[500,550,800]
plt.xlabel("year")
plt.ylabel("population")
plt.bar(x,y)
```

```python
age=[5,7,9,12,13,15,14,21,32,53,54,66,78,54,44]
bin=[0,10,20,30,40,50,60,70,80]
weight=[1,2,5,10,11,21,22,31,41,51,61,71]
plt.hist(age,bin)
plt.hist(weight,bin)
plt.xlabel("height")
plt.ylabel("age")
plt.title("Age Vs Height Vs weight")
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,8,7,4,5,6,3,2,1]
plt.scatter(x,y,label="Some Label",color="red")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("Scatter Plots")
plt.legend()
plt.show()
```

```python
days=[1,2,3,4,5]
sleeping=[10,12,11,10,11]
studying=[1,0,1,0,1]
playing=[5,5,5,5,5]
plt.stackplot(days,sleeping,studying,playing)
plt.xlabel("Time")
plt.ylabel("Work")
plt.title("My Day TT")
plt.show()
```

```python
chart=[12,5,0.1,5]
label=["Sleeping","playing","studying","eating"]
plt.pie(chart,labels=label)
plt.title("Titetable of an IITian")
plt.show()
```
