# seaborn

```python
from matplotlib import pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset('iris')
sb.histplot(df['petal_length'],kde=True)
plt.show()
```

```python
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'],hist=False)
plt.show()
```

```python
sb.jointplot(x='petal_length',y='petal_width',data=df)
plt.show()
```

```python
sb.jointplot(x='petal_length',y='petal_width',data=df,kind='hex')
plt.show()
```

```python
sb.pairplot(df,hue='species',diag_kind='kde',kind='scatter',palette='bright')
plt.show()
```

```python
sb.boxplot(data=df,orient='h')
plt.show()
```

```python
sb.violinplot(x='species',y='petal_length',data=df)
plt.show()
```

```python
g=sb.PairGrid(df)
g.map(plt.scatter)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[float, int])
corr=numeric_df.corr()
sb.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```
