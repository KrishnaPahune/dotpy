## Printing Mean,Max,Std,Min and Count
this function is used to print the mean, max, min, standard deviation and count of the selected features. It uses for loop

```python
def info():
    for i in range(0,4):
        print(f"The mean of {columns[i]} is {df[columns[i]].mean()}")
        print(f"The count of {columns[i]} is {df[columns[i]].count()}")
        print(f"The stanard deviation of {columns[i]} is {df[columns[i]].std()}")
        print(f"The minimum of {columns[i]} is {df[columns[i]].min()}")
        print(f"The maximum of {columns[i]} is {df[columns[i]].max()}")
        print("-".center(50,"-"))
```
