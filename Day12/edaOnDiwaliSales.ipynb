```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
df=pd.read_csv('Diwali.csv')
print(df.to_string())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
new_df=df.fillna(df['Amount'].mean(),inplace=False)
print(new_df.isnull().sum())
```

```python
duplicates=new_df.duplicated().sum()
duplicates
```

```python
new_df=new_df.drop_duplicates()
duplicates=new_df.duplicated().sum()
print(duplicates)
```

```python
new_df
```

```python
new_df=new_df.sort_values('Cust_name')
new_df
```

```python
#pie chart for zone counts
zone_counts=new_df['Zone'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(zone_counts,labels=zone_counts.index,autopct='%1.1f%%',startangle=140)
plt.title('Distribution of Zones')
plt.show()
```

```python
#pie chart for occupation counts
occupation_counts=new_df['Occupation'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(occupation_counts,labels=occupation_counts.index,autopct='%1.1f%%',startangle=140)
plt.title('Distribution of Occupations')
plt.show()
```

```python
#Plotting the bar chart for gender and its count
plt.figure(figsize=(5,5))
sb.countplot(x='Gender', data=new_df, hue='Gender', palette=['black', 'red'])
plt.title('Count of Each Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sb.histplot(data=new_df,x='Age Group',hue='Gender',multiple='stack',bins=10,kde=True,palette=['black','red'])
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()             
```

```python
#total no of orders from top 10 states
top_states = new_df['State'].value_counts().index[:10]
order_counts = new_df['State'].value_counts().values[:10]

plt.figure(figsize=(8,8))
plt.pie(order_counts, labels=top_states, autopct='%1.2f%%', startangle=140)
plt.title('Total Number of Orders from Top 10 States')
plt.show()
```

```python
Top_10=new_df.groupby(['Amount','Occupation','State']).size().reset_index()
Top_10
```

```python
# COunt the most common occupation
occupation_counts=new_df['Occupation'].value_counts()
most_common_occupation = occupation_counts.idxmax()
num_buyers_in_occupation = occupation_counts.max()
print(f"The occupation with the most no of counts is {most_common_occupation} with number of buyers is {num_buyers_in_occupation}")
```

```python
# COunt the most common age group
age_group_counts=new_df['Age Group'].value_counts()
most_common_age_group = age_group_counts.idxmax()
num_buyers_in_age_group = age_group_counts.max()
print(f"The age group with the most no of counts is {most_common_age_group} with number of buyers is {num_buyers_in_age_group}")
```

```python
# calculate the average purchase amount for each occupation
avg_purchase_by_occupation = df.groupby('Occupation')['Amount'].mean()
print('Avg purchase amount by occupation:')
print(avg_purchase_by_occupation)
```

```python
print(avg_purchase_by_occupation.max())
```

```python
# calculate the total sales by age group
total_sales_by_age_group = new_df.groupby('Age Group')['Amount'].sum().astype(int)
print(' Total sales amount by age group:')
print(total_sales_by_age_group)
```

```python
# Count the number of male and female buyers
gender_counts=new_df['Gender'].value_counts()
most_common_gender=gender_counts.idxmax()
num_buyers_in_gender=gender_counts.max()
print(f"The most common gender among buyers is {most_common_gender} having a count of {num_buyers_in_gender}")
```

```python
# Count the most common marital status
marital_status_count=new_df['Marital_Status'].value_counts()
most_common_marital_status=marital_status_count.idxmax()
num_buyers_in_marital_status=marital_status_count.max()
print(f"The most common marital status among buyers is {most_common_marital_status} having a count of {num_buyers_in_marital_status}")
```
